```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("ASL762_MarksMinor2.xlsx")
```

```python
df.head()
```

```python
df.columns
```

```python
# rename columns  to make them more readable

df.rename(columns={"Student Name (ASL762)" :"name", "Minor-1 (*/25)" : "minor_1", "Minor-2 (*/25)" : "minor_2", }, inplace=True)
```

```python
df.head()
```

```python
# add total marks column
df["total"] = df["minor_1"] + df["minor_2"]
```

```python
df.head()
```

```python
df.describe()
```

```python
df
```

```python
# display the Name of the Nan values in Total column
df[df["total"].isna()]
```

```python
# drop Nan values
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
df
```

```python
# plot the distribution of total marks
def plot_distribution(df):
    plt.figure(figsize=(15, 10), dpi=250)
    sns.displot(df["total"], kde=False, bins=20)
    plt.xlabel("Total Marks")
    plt.ylabel("Frequency")
    plt.title("Distribution of Total Marks")
    plt.show()

plot_distribution(df)
```

```python
# plot density plot of minor_1 and minor_2 marks
def plot_density(df):
    plt.figure(figsize=(15, 10), dpi=250)
    sns.kdeplot(df["minor_1"], fill=True, color="red")
    sns.kdeplot(df["minor_2"], fill=True, color="blue")
    sns.kdeplot(df["total"], fill=True, color="yellow")
    plt.xlabel("Marks")
    plt.ylabel("Density")
    plt.title("Density Plot of Minor 1 and Minor 2 Marks")
    plt.show()

plot_density(df)
```
