```python
import numpy as np
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
```

# Hard Steel

```python
# Print Out Table
readings = pd.DataFrame()
readings['Mag_Field (Guass)'] = [ 33, 67, 107, 153, 198, 248, 296, 344, 384 ] # Gauss
readings['2*loop_width (mV)'] = [ 50, 97.5, 108.75, 112.5, 115.75,116.25 , 117.5, 118.75, 118.75 ] # mV
readings['2*loop_width (mm)'] = readings['2*loop_width (mV)']/10 #mm
readings['tip_to_tip_width (mV)'] = [ 268.25, 822.75, 1022.25, 1109.25, 1156.25, 1192.75, 1221.50, 1239.75, 1250.75 ] # mV
readings['2*Intercept (mV)'] = [ 117.50, 540.00, 616.25, 634.25, 634.25, 638, 634.25, 634.25, 634.25  ] # mV
readings
```

```python
readings = pd.DataFrame()
readings['Mag_Field'] = [ 33, 67, 107, 153, 198, 248, 296, 344, 384 ] # Gauss
readings['2*loop_width'] = [ 50, 97.5, 108.75, 112.5, 115.75,116.25 , 117.5, 118.75, 118.75 ] # mV
readings['2*loop_width'] = readings['2*loop_width']/10 #mm
readings['tip_to_tip_width'] = [ 268.25, 822.75, 1022.25, 1109.25, 1156.25, 1192.75, 1221.50, 1239.75, 1250.75 ] # mV
readings['2*Intercept'] = [ 117.50, 540.00, 616.25, 634.25, 634.25, 638, 634.25, 634.25, 634.25  ] # mV
readings
```

```python
fitted_polynomial_coeff = np.polyfit(readings['Mag_Field'],readings['2*loop_width'],12)
fitted_polynomial = np.poly1d(fitted_polynomial_coeff)
fitted_data = pd.DataFrame()
fitted_data['x'] = np.linspace(25,395,num=2000)
fitted_data['y'] = fitted_polynomial(fitted_data['x'])

# straight fitted line
X = [ 248, 296, 344, 384 ] # Gauss
Y = [ 11.625, 11.750, 11.875	, 11.875] # mm
fitted_polynomial_coeff = np.polyfit(X,Y,1)
fitted_polynomial = np.poly1d(fitted_polynomial_coeff)
fitted_data_sl = pd.DataFrame()
fitted_data_sl['x'] = np.linspace(0,395,num=2000)
fitted_data_sl['y'] = fitted_polynomial(fitted_data['x'])
#

# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=readings['Mag_Field'], y=readings['2*loop_width'],mode='markers',name='Observations'))
fig.add_trace(go.Scatter(x=fitted_data['x'], y=fitted_data['y'],
                    mode='lines',
                    name='Fitted Curve'))

a1 = fitted_data_sl['x'][0]
a2 = fitted_data_sl['y'][0]
print("a2 = ", a2)
fig.add_hline(y=a2,annotation_text="y=8.590191804281847",annotation_position='top',line_dash='dash')


fig.add_trace(go.Scatter(x=fitted_data_sl['x'], y=fitted_data_sl['y'],
                    mode='lines',
                    name='Fitted Straight Line'))
fig.update_layout(xaxis_title="Magnetic Field (Gauss)",yaxis_title="2L*oop Width (mm)",height=600,title = 'Magnetic Field (Gauss) vs 2*Loop Width  (Hard Steel)')
```

```python
# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=readings['Mag_Field'], y=readings['2*Intercept'],mode='markers+lines',name='Observations'))
fig.add_hline(y=643.25,annotation_text="y=643.25",annotation_position='top')
fig.update_layout(xaxis_title="Magnetic Field (Gauss)",yaxis_title="2*Intercept (mV)",width=1000,height=700,
                  title = ' Magnetic Field (Gauss) vs 2*Intercept (Hard Steel)')
```

```python
# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=readings['Mag_Field'], y=readings['tip_to_tip_width'],mode='markers+lines',name='Observations'))
fig.add_hline(y=1265,annotation_text="y=1265",annotation_position='top')
fig.update_layout(xaxis_title="Magnetic Field (Gauss)",yaxis_title='Tip to Tip Width (mV)',height=700,
                  title = ' Tip to Tip Width vs Magnetic Field Graph (Hard Steel)')
```
