```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] =(10,6)
plt.rcParams['figure.dpi'] = 150
import pandas as pd
import seaborn as sns
from scipy.optimize import curve_fit
```

```python
res_160 = pd.read_csv('res/aka-res-160-data.csv')
res_181 = pd.read_csv('res/aka-res-181-data.csv')
res_213 = pd.read_csv('res/aka-res-213-data.csv')
res_242 = pd.read_csv('res/aka-res-242-data.csv')
res_270 = pd.read_csv('res/aka-res-270-data.csv')
res_300 = pd.read_csv('res/aka-res-300-data.csv')
```

```python
def plot_res(data,x_data=' Voltage(V)',y_data='Current (A)'):
    def straight_line(x,m,c):
        return m*x + c

    popt, pcov = curve_fit(straight_line, data[x_data], data[y_data], p0=(1,1))

    x = np.linspace(-10,10,100)
    y = straight_line(x,popt[0],popt[1])
    plt.plot(x,y)
    plt.xlabel('Voltage(V)')
    plt.ylabel("Current(A)")
    return 1/popt[0]
```

```python
slope = []

for i in [res_160,res_181,res_213,res_242,res_270,res_300]:
    slope.append(plot_res(i))
```

```python
slope
```

```python
temp = [160,181,213,242,270,300]
def temp_res(x,a,b,c):
    return a*x**2 + b*x +c

pcoe, pcov = curve_fit(temp_res,temp,slope,p0=(1,1,10))
x = np.linspace(160,300,100)
y = temp_res(x,pcoe[0],pcoe[1],pcoe[2])
plt.plot(x,y)
plt.scatter(temp,slope);
plt.xlabel('Temperature(K)')
plt.ylabel('Resistance(Ohm)')
```
