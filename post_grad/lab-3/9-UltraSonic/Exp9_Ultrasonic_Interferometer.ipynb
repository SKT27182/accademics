```python
ydata = readings['Current (uA)']
xdata = readings['Total (mm)']

plt.figure(figsize=(12, 6),dpi=150)
plt.plot(xdata, ydata, '*', label='data')
plt.xticks(np.arange(0, 3.4, 0.1));
plt.yticks(np.arange(0, 40,2));


plt.grid(linestyle = '--', linewidth = 0.5)
plt.xlim(0,3.2)
plt.ylim(0,40)
plt.xlabel('Length (mm)')
plt.ylabel('Current (uA)');
plt.title('Current vs Length')
```

```python
# Plot Scatter Plot here

trace = go.Scatter(x=readings['Total (mm)'], y=readings['Current (uA)'], mode='markers')

data = [trace]

layout = go.Layout(title="Current (micro-Amp) vs Length (milimeter) | Liquid ", xaxis={'title':'Length (mili-meter)'}, yaxis={'title':'Current (micro-Amp)'})

fig= go.Figure(data=data, layout=layout)

#pyo.plot(fig)
fig.show()
```

```python
x = readings['Total (mm)']
y = readings['Current (uA)']

# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='markers', name='Observations'))

from scipy import signal 
import scipy
yy = signal.savgol_filter( y, 9, 5) # https://en.wikipedia.org/wiki/Savitzky%E2%80%93Golay_filter
fig.add_trace(go.Scatter(x=x, y=yy, mode='lines', name='Smoothing'))
fig.update_layout(width=1500, height=800 ,legend=dict(yanchor="top", y=0.99, xanchor="left", x=0.01 ), font=dict(size=17.5),
                    xaxis_title='Length (mili-meter)', yaxis_title='Current (micro-Amp)',
                    title="Current (micro-Amp) vs Length (milimeter) | Liquid ",)

fig.add_vline(x=0.495,annotation_text="x=0.495", line_width=2, line_dash="dash", line_color="green")
fig.add_vline(x=1.25,annotation_text="x=1.25" ,line_width=2, line_dash="dash", line_color="green")
fig.add_vline(x=2.0,annotation_text="x=2.0", line_width=2, line_dash="dash", line_color="green")
fig.add_vline(x=2.725,annotation_text="x=2.725", line_width=2, line_dash="dash", line_color="green")

```
