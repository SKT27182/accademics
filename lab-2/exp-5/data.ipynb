<B><h1> distance = 10mm------------------------------------------------------------------------</B>

```python
data10 = pd.DataFrame()
data10["Voltage_V"] = [60,70,80,90,100,110,120,130,140,150,160]
data10["Current_A"] = [0.02 ,0.06 ,0.08 ,0.15 , 0.18, 0.25, 0.30, 0.36, 0.42, 0.47, 0.56]
data10["Temperature_C"] = [35.3, 36.8, 37.6, 39.7, 41.2, 42.5, 43.2, 44.9, 45.7, 46.6, 48.1]
data10
```

```python
plt.figure(figsize=(10,6),dpi=150)
plt.plot(data10["Current_A"],data10['Temperature_C'])
plt.xlabel("Current(A)")
plt.ylabel("Plume Temperature in degree Celcius")
plt.title("Temperature Profile of Pluume length at fixed rate with Current")
sns.scatterplot(x=data10["Current_A"],y=data10['Temperature_C'])
```
