```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#Function for which we want to fit the curve
def func(x,m,c):
    return m*x + c
```

<B>For frequency= 14.01 MHz and P=152.4 V</B>

```python
readings1 = pd.DataFrame()
readings1["Current I(mA)"]= [104,132,160,187,215,241]       
readings1["2Q (V)"] = [97, 75.4, 63, 53, 45.8, 41.4]
readings1["Current I(A)"] = readings1["Current I(mA)"]/1000 
readings1["Q (cm)"] = readings1["2Q (V)"]/(16*2)
readings1["1/I (A^-1)"] = 1/readings1["Current I(A)"]
readings1
```

```python
#fitting the curve by providing function, xdata,ydata and boundaries and it will return the fitted constants of the given function in popt
popt, pcov = curve_fit(func,readings1["1/I (A^-1)"],readings1["Q (cm)"],bounds=(0,9))
#popt contains the list of the constants 
funcdata = func(np.linspace(0,10,30),popt[0],popt[1])   #functdata contains the fitted ydata , here we can provide any xdata as we are now generating ydata on the fitted curve function
plt.figure(figsize=(8,4),dpi=150)
#Ploting the x data(our original data) and y(our original data)
plt.plot(readings1["1/I (A^-1)"],readings1["Q (cm)"], 'bo',label='experimental-data',lw=2)
#Ploting the fitted curve for and random x
plt.plot(np.linspace(0,10,30),funcdata,'r',label="Fitted-line",lw=0.7);
plt.xlim(0,10)
plt.ylim(0,5)
plt.legend()
plt.ylabel("1/I(A^-1)")
plt.xlabel("Q(cm)")
popt
```
