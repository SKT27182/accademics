```python
grading_system = {"A":10, "A-":9, "B":8, "B-":7, "C":6, "C-":5, "D":4, "D-":3, "E":2, "F":1, "F-":0}
```

```python
while(input("Want to enter data: ") == "yes"):
    subject = input("Enter subject: ")
    course_type = input("Enter course type: ")
    grade = input("Enter grade: ")
    credit = int(input("Enter credit: "))
    cgpa.loc[len(cgpa)] = [subject, course_type, grade, credit]
```

```python
cgpa
```

```python
sem_1 = cgpa.iloc[:5,:]
sem_2 = cgpa.iloc[5:13,:]
sem_3 = cgpa.iloc[13:,:]
sem_4 = cgpa.iloc[18:,:]
```

```python
print("Semester-1\n\n",sem_1,"\n\nSemester-2\n\n",sem_2, "\n\nSemester-3\n\n",sem_3, "\n\nSemester-4\n\n",sem_4)
```

```python
def calculate_cgpa(cgpa):
    total_credit = 0
    total_grade = 0
    for i in range(len(cgpa)):
        if cgpa.iloc[i][2] != "Audit":
            total_credit += cgpa.iloc[i][3]
            total_grade += cgpa.iloc[i][3] * grading_system[cgpa.iloc[i][2]]
    return total_grade/total_credit
```

```python
print("Semester-1: ",calculate_cgpa(sem_1))
print("Semester-2: ",calculate_cgpa(sem_2))
print("Semester-3: ",calculate_cgpa(sem_3))
print("Semester-4: ",calculate_cgpa(sem_4))
print("Total:      ",calculate_cgpa(cgpa))
```
